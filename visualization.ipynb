# Visualization

```python
import json
import matplotlib.pyplot as plt

# Load the JSON data from file
with open("data/metric_summary_results.json", "r") as file:
    data = json.load(file)

# Convert the JSON data to the desired format
metrics = {
    "Average Exact Match": data["average_exact_match"],
    "Average BLEU": data["average_bleu"],
    "Average CHRF": data["average_chrf"],
    "Functional Accuracy": data["average_functional_accuracy"]
}

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(metrics.keys(), metrics.values(), color=['skyblue', 'salmon', 'lightgreen'])
plt.ylim(0, 1)  # Since metrics are scores between 0 and 1

# Adding labels
plt.xlabel("Metrics")
plt.ylabel("Scores")
plt.title("Evaluation of tiny_startcoder_py model for middle code completion")
plt.xticks(rotation=45)
for index, value in enumerate(metrics.values()):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom')

# Display the plot
plt.tight_layout()
plt.show()

plt.savefig("artifacts/results.png")
```
